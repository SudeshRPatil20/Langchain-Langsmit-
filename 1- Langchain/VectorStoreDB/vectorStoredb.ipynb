### FAISS

Fecebook AI Similarity Search is a library which is used for ifficient search and clustering algorithms. it contains algorithms that can search in set of vectors that of any size up to possible that does not fit in any RAM.therefore they are used in Parameter tuning and evalution.

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader=TextLoader("speech.txt")
documents=loader.load()
text_splitter=CharacterTextSplitter(chunk_size=1000, chunk_overlap=30)
docs=text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model="gemma:2b")               # Initializes the embedding model using Ollama
db = FAISS.from_documents(docs, embeddings)   # Creates a FAISS vector store from the documents
db               
```

```python
query="how does speech must be planed?"
docs=db.similarity_search(query)
docs[0].page_content
```

```python
# this is one way of doing query now let us see another way of doing query.
# Retriver 
# it is class used to concetc vectorstore with langchain . it allows us to use other langchain methods works largely with retriver

retriver=db.as_retriever()
docs=retriver.invoke(query)
docs[0].page_content
 # as you can see that retriver are able to do similar kind of work as above but when you use it for higher projects you need to first convert that vectorstoreDB to retriver that use it.
```

### SIMILARITY SEARCH
It is kind of method of fiss that is used to find the similarity search score of query with retrive documents. lower the document highter the right answer

```python
similarity_search_with_score=db.similarity_search_with_score(query)
similarity_search_with_score
```

```python
# we can directly pass embedding vectors for search like query as follows
embedding=embeddings.embed_query(query)
embedding
```

```python
docs_And_score=db.similarity_search_by_vector(embedding)
```
